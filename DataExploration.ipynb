```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/data.csv')

df.drop('EmpID', axis = 1, inplace = True)
df.drop('Emp Name', axis = 1, inplace = True)
df.drop('Attrition ', axis = 1,inplace = True)
df
```

```python
df.loc[df["Grade"] == 'E1', "Grade"]= 1
df.loc[df["Grade"] == 'E2', "Grade"]= 2
df.loc[df["Grade"] == 'M1', "Grade"]= 3
df.loc[df["Grade"] == 'M2', "Grade"]= 4
df.loc[df["Grade"] == 'M3', "Grade"]= 5
df.loc[df["Grade"] == 'M4', "Grade"]= 6
df.loc[df["Grade"] == 'CX0', "Grade"]= 7
df['Tenure'] = df['Tenure'].str.replace(r'\s+', '', regex=True)  # Remove extra spaces
df['Marital Status'] = df['Marital Status'].str.replace(r'\s+', '', regex=True)  # Remove extra spaces
df['Education'] = df['Education'].str.replace(r'\s+', '', regex=True)  # Remove extra spaces


df['Tenure'] = df['Tenure'].astype(float)

df
```

```python
x = df['S.No'].values
y = df['Tenure'].values

# Calculate the minimum and maximum values of 'Tenure'
min_tenure = np.min(y)
max_tenure = np.max(y)

# Generate evenly spaced y-axis ticks
y_ticks = np.linspace(min_tenure, max_tenure, num=9)  # Adjust 'num' as needed

# Create the scatter plot
plt.plot(x, y, 'ro')

# Set the y-axis ticks
plt.yticks(y_ticks)

# Labeling axes and adding a title
plt.ylabel('Tenure')
plt.xlabel('S.No')
plt.title('Tenure of employees left')

# Show the plot
plt.grid(True)  # Optional: Add gridlines
print(df['Tenure'].describe())
plt.show()
```
